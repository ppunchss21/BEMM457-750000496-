```
import pandas as pd

df_2023 = pd.read_csv('/content/2023 W1 spring-Central.csv')
df_2024 = pd.read_csv('/content/2024 W1 spring-Central.csv')

combined_df = pd.concat([df_2023, df_2024], ignore_index=True)

combined_df.to_csv('/content/combined_data.csv', index=False)
print("Combined data saved to /content/combined_data.csv")
```
